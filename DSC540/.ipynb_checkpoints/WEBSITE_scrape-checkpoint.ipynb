```python
# import libraries

from bs4 import BeautifulSoup as bs
import pandas as pd
import csv
import matplotlib
import numpy as np
import pandas as pd
import requests as rq
from scipy import stats
```

```python
#

url="https://fantasydata.com/nfl/adp?season=2023&leaguetype=1"
html_content = rq.get(url).text
soup=bs(html_content, 'html.parser')
print(soup.prettify())

```

```python
#

gdp=soup.find("table", attrs={"class":"wikitable"})
data=gdp.tbody.find_all("tr")
data


```

```python
#

all_tables = soup.find_all("table")
print("Total number of tables are {} ".format(len(all_tables)))

```

```python
data_table = soup.find("table")
print(type(data_table))
```

```python
sources = data_table.tbody.findAll('tr', recursive=False)[1]
sources_list = [td for td in sources.findAll('td')]
print(len(sources_list))
```
