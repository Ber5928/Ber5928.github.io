```python
# import libraries

import pandas as pd
```

```python
# load the CSV file into a data frame

file_path = '/Users/brianreppeto/DSC540/teams.csv'
teams_df = pd.read_csv(file_path)
```

```python
# head the new data frame to see file

teams_df.head()
```

#### Step 1

```python
# Remove unneeded columns

columns_to_remove = ["sbr", "sbr_wins", "sbr_name", "draft_kings"]
teams_df = teams_df.drop(columns=columns_to_remove, errors='ignore')
```

```python
# head the new data frame to see removed columns

teams_df.head()
```

#### Step 2

```python
# Filter rows where the "season" is greater than or equal to 2017

teams_df = teams_df[teams_df['season'] > 2017]
```

```python
# head the data frame to see removed rows 

teams_df.head()
```

#### Step 3

```python
# make all the column headings in upper letters

teams_df.columns= teams_df.columns.str.upper()
```

```python
# head the data frame to see the changed upper column headings 

teams_df.head()
```

#### Step 4

```python
# Create a new column "TEAM_SEASON" and fill it with the concatenation of "SEASON" and "TEAM" columns

teams_df['TEAM_SEASON'] = ''

for index, row in teams_df.iterrows():
    teams_df.at[index, 'TEAM_SEASON'] = str(row['SEASON']) + '_' + str(row['TEAM'])
```

```python
# head the data frame to see the updated DF with the new column

teams_df.head()
```

#### Step 5

```python
# Capitalize the first letter of each word in the hyphenated column

teams_df['HYPHENATED'] = teams_df['HYPHENATED'].str.title()
```

```python
# head the data frame to see the updated DF with the new hyphenated upper case letter for each word

teams_df.head()
```

### Step 6

```python
# Save the cleaned data to a new CSV file

output_file_path = '/Users/brianreppeto/DSC540/teams_data_cleaned.csv'
teams_df.to_csv(output_file_path, index=False)
```
