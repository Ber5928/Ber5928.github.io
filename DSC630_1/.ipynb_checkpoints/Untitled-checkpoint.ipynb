```python
import pandas as pd

# Load the dataset provided by the user
file_path = 'dodgers-2022.csv'
dodgers_data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
dodgers_data.head()
```

```python
import matplotlib.pyplot as plt

# Convert 'attend' to numeric in case there are any formatting issues
dodgers_data['attend'] = pd.to_numeric(dodgers_data['attend'], errors='coerce')

# Group by day of the week to analyze attendance patterns
attendance_by_day = dodgers_data.groupby('day_of_week')['attend'].mean().sort_values(ascending=False)

# Group by promotional events to analyze attendance patterns
attendance_by_promotions = dodgers_data.groupby(['cap', 'shirt', 'fireworks', 'bobblehead'])['attend'].mean()

# Plot attendance by day of the week
plt.figure(figsize=(10, 6))
attendance_by_day.plot(kind='bar', color='skyblue')
plt.title('Average Attendance by Day of the Week')
plt.ylabel('Average Attendance')
plt.xlabel('Day of the Week')
plt.xticks(rotation=45)
plt.show()

# Display the attendance by promotional events to the user
import ace_tools as tools; tools.display_dataframe_to_user(name="Attendance by Promotions", dataframe=attendance_by_promotions)

attendance_by_day, attendance_by_promotions
```
