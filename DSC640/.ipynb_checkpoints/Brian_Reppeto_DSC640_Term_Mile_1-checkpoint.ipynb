```python
import pandas as pd
import matplotlib.pyplot as plt



# Load the uploaded Excel file to explore its content
file_path = 'nationaldatabaseofchildcareprices.xlsx'
excel_data = pd.ExcelFile(file_path)

# Display sheet names to understand the structure of the file
excel_data.sheet_names
```

```python
# Load the main sheet to examine the content
data = excel_data.parse('nationaldatabaseofchildcare')

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python

# selecting relevant columns for analysis
selected_columns = [
    "State_Name", 
    "County_Name", 
    "StudyYear", 
    "_75FCCInfant", 
    "_75FCCToddler", 
    "_75FCCPreschool"
]

# Filtering dataset for relevant columns and dropping missing values
filtered_data = data[selected_columns].dropna()

# Calculating top 10 and bottom 10 states for infant care costs
state_avg_cost = filtered_data.groupby("State_Name")["_75FCCInfant"].mean().sort_values(ascending=False).head(10)
bottom_10_states = filtered_data.groupby("State_Name")["_75FCCInfant"].mean().sort_values(ascending=True).head(10)

# Resetting indices for side-by-side plotting
top_states = state_avg_cost.reset_index()
bottom_states = bottom_10_states.reset_index()

# Simplifying the chart with labels for Bottom 10 States
plt.figure(figsize=(14, 8))
bar_width = 0.4
x_top = range(len(top_states))
x_bottom = [x + bar_width for x in x_top]

# Plotting bars for Top 10 and Bottom 10 states
plt.bar(x_top, top_states["_75FCCInfant"], width=bar_width, color='dodgerblue', edgecolor='black', label="Top 10 States")
plt.bar(x_bottom, bottom_states["_75FCCInfant"], width=bar_width, color='orange', edgecolor='black', label="Bottom 10 States")

# Adding labels above the orange bars (Bottom 10 States)
for i, value in enumerate(bottom_states["_75FCCInfant"]):
    if pd.notna(value):  # Avoid labeling missing values
        plt.text(x_bottom[i], value + 1, str(bottom_states["State_Name"][i]), ha='center', fontsize=10, color='black')

# Finalizing chart details
plt.title("Comparison of Top 10 vs Bottom 10 States for Average Infant Care Costs", fontsize=16)
plt.xlabel("States", fontsize=14)
plt.ylabel("Average Cost ($)", fontsize=14)
plt.xticks([x + bar_width / 2 for x in x_top], top_states["State_Name"], rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12, title="State Groups")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```

```python
# Checking if the "MHI" column exists and contains data
if "MHI" in data.columns:
    # Filtering for states with median household income data and dropping missing values
    filtered_income_data = data[["State_Name", "MHI"]].dropna()
    
    # Calculating top 10 and bottom 10 states for median household income
    top_income_states = filtered_income_data.groupby("State_Name")["MHI"].mean().sort_values(ascending=False).head(10)
    bottom_income_states = filtered_income_data.groupby("State_Name")["MHI"].mean().sort_values(ascending=True).head(10)
    
    # Resetting indices for side-by-side plotting
    top_income_states = top_income_states.reset_index()
    bottom_income_states = bottom_income_states.reset_index()
    
    # Simplifying the chart with labels for Bottom 10 States
    plt.figure(figsize=(14, 8))
    bar_width = 0.4
    x_top_income = range(len(top_income_states))
    x_bottom_income = [x + bar_width for x in x_top_income]
    
    # Plotting bars for Top 10 and Bottom 10 states
    plt.bar(x_top_income, top_income_states["MHI"], width=bar_width, color='dodgerblue', edgecolor='black', label="Top 10 States")
    plt.bar(x_bottom_income, bottom_income_states["MHI"], width=bar_width, color='orange', edgecolor='black', label="Bottom 10 States")
    
    # Adding labels above the orange bars (Bottom 10 States)
    for i, value in enumerate(bottom_income_states["MHI"]):
        if pd.notna(value):  # Avoid labeling missing values
            plt.text(x_bottom_income[i], value + 1000, str(bottom_income_states["State_Name"][i]), ha='center', fontsize=10, color='black')
    
    # Finalizing chart details
    plt.title("Comparison of Top 10 vs Bottom 10 States for Median Household Income", fontsize=16)
    plt.xlabel("States", fontsize=14)
    plt.ylabel("Median Household Income ($)", fontsize=14)
    plt.xticks([x + bar_width / 2 for x in x_top_income], top_income_states["State_Name"], rotation=45, ha='right', fontsize=12)
    plt.yticks(fontsize=12)
    plt.legend(fontsize=12, title="State Groups")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("The column 'MHI' (Median Household Income) does not exist in the dataset.")
```

```python
# Adding a column for childcare cost as a percentage of Median Household Income (MHI)
if "MHI" in data.columns:
    # Filtering relevant columns and dropping missing values
    relative_cost_data = data[["State_Name", "_75FCCInfant", "MHI"]].dropna()
    
    # Calculating childcare cost as a percentage of MHI
    relative_cost_data["Cost_Percentage_MHI"] = (relative_cost_data["_75FCCInfant"] / relative_cost_data["MHI"]) * 100
    
    # Aggregating by state to find average percentage cost
    relative_cost_by_state = relative_cost_data.groupby("State_Name")["Cost_Percentage_MHI"].mean().sort_values(ascending=False)
    
    # Plotting the states with the highest childcare costs relative to MHI
    top_relative_cost_states = relative_cost_by_state.head(10)
    plt.figure(figsize=(12, 7))
    top_relative_cost_states.plot(kind='bar', color='lightcoral', edgecolor='black')
    plt.title("Top 10 States with Highest Childcare Costs Relative to Median Household Income", fontsize=16)
    plt.xlabel("State", fontsize=14)
    plt.ylabel("Cost as Percentage of MHI (%)", fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("The column 'MHI' (Median Household Income) does not exist in the dataset.")
```

```python
# Fixing and refining the code for temporal trends and age-specific costs

# Recalculating the mean costs by year for temporal trends
mean_costs_by_year = filtered_data.groupby("StudyYear")[
    ["_75FCCInfant", "_75FCCToddler", "_75FCCPreschool"]
].mean()

# Refining the temporal trends chart
plt.figure(figsize=(12, 7))
for column in mean_costs_by_year.columns:
    plt.plot(mean_costs_by_year.index, mean_costs_by_year[column], label=column.replace("_", " "), linewidth=2)

plt.title("National Average Childcare Costs Over Time", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Average Cost ($)", fontsize=14)
plt.legend(title="Childcare Type", fontsize=12, title_fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Recalculating age group means for age-specific costs
age_group_means = filtered_data[["_75FCCInfant", "_75FCCToddler", "_75FCCPreschool"]].mean()

# Refining the age-specific costs chart
plt.figure(figsize=(10, 6))
age_group_means.plot(kind='bar', color=['tomato', 'gold', 'mediumseagreen'], edgecolor='black')
plt.title("National Average Childcare Costs by Age Group", fontsize=16)
plt.xlabel("Childcare Type", fontsize=14)
plt.ylabel("Average Cost ($)", fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Importing the required library for the heatmap
import seaborn as sns

# Reattempting to plot the correlation matrix
if "MHI" in data.columns:
    # Selecting the relevant columns for correlation analysis
    correlation_data = data[["_75FCCInfant", "_75FCCToddler", "_75FCCPreschool", "MHI"]].dropna()
    
    # Calculating the correlation matrix
    correlation_matrix = correlation_data.corr()
    
    # Plotting the correlation matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title("Correlation Between Median Household Income (MHI) and Childcare Costs", fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()
else:
    print("The column 'MHI' (Median Household Income) is not available in the dataset.")
```
