## Activity 6 Working with the Adult Income Dataset

```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read the adult_income_data. dataset

adult_df=pd.read_csv("adult_income_data.csv")
```

```python
# read the first 10 rows

adult_df.head(10)
```

```python
# using the .txt file  create a empty list, open the .txt file loop thru each line in .txt seperated by ':'.
# append to the headers list

headers = []  # initializes an empty list called headers to store the extracted headers.
with open('adult_income_names.txt','r') as b:
    for line in b:
        b.readline()
        var=line.split(":")[0]
        headers.append(var)
```

```python
# Output the headers list values

headers
```

```python
# append the 'Income' name to the response dataset

headers.append('Income')
```

```python
# update the adult_df with the new names

adult_df=pd.read_csv("adult_income_data.csv",names=headers)

# Output the updated file for the new column

adult_df.head(10)
```

```python
# Missing data for adult_income_data DF 

adult_df.isnull().sum()
```

```python
# Create a subset df with only 'age', 'education', and 'occupation'.

adult_sub=adult_df[['age','education','occupation']]
```

```python
# Head the new subset df to see changes

adult_sub.head(10)
```

```python
# Plot a histogram of age with a bin size of 20

adult_sub['age'].hist(bins=20)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Function to strip whitespace characters using the python strip method command

def stp_wht_spc(c):
    return c.strip()
```

```python
# Use the apply method to apply the stp_wht_spc function to the adult_sub df string columns

# Education columns

adult_sub['ed_strip']=adult_df['education'].apply(stp_wht_spc)
adult_sub['education']=adult_sub['ed_strip']
adult_sub.drop(labels=['ed_strip'],axis=1,inplace=True)

# Occupation column

adult_sub['oc_strip']=adult_df['occupation'].apply(stp_wht_spc)
adult_sub['occupation']=adult_sub['oc_strip']
adult_sub.drop(labels=['oc_strip'],axis=1,inplace=True)

```

```python
# Head the subset df

adult_sub.head(10)
```

```python
# Find the number of people who are aged between 30 and 50

adult_ages=adult_sub[(adult_sub['age']>=30) & (adult_sub['age']<=50)]
```

```python
# Head the filter to show the results

adult_ages.head(10)
```

```python
# Create a subset df with only 'age', 'education' to use for the group by below

adult_sub1=adult_sub[['education','age']]
```

```python
# Create a subset df people who are aged between 30 and 50

adult_sub2=adult_sub1[(adult_sub['age']>=30) & (adult_sub['age']<=50)]
```

```python
# head the new DF to see results

adult_sub2.head(10)
```

```python
# Group the record based on age and education to find the mean age
# Since the find the number between 30 and 50 was the question before this I assumed this was the ask
# If my assumption was wrong I can change, just let me know

adult_sub2.groupby(['education']).mean()
```

```python
# Group by occupation and show summary stats for age

result=adult_sub.groupby('occupation').describe()['age'] 


# Sort by mean age in descending order
result_sorted = result.sort_values(by='mean', ascending=False)

# Display the sorted result
print(result_sorted)
```

The profession with the oldest workers on average is "Exec-managerial".
The profession that has the largest share above 75% is the "?" profession.

```python
# Use subset and groupby to find the outliers

out_stat=adult_sub.groupby('occupation').describe()['age']

# Sort by count in descending order
out_sorted = out_stat.sort_values(by='count', ascending=False)

# Display the sorted result
print(out_sorted)
```

The Armed-Forces occucation appears to be an outlier.


```python
# Plot the values of the bar chart

plt.figure(figsize=(13,10))
plt.barh(y=out_stat.index,width=out_stat['count'])
plt.yticks(fontsize=15)
plt.show()
```

```python
# Merge data using common keys.
# create new df by subset of columns from orig df then sample 10 rows and seed the rand # gen

merge_adult=adult_df[['age','workclass', 'occupation']].sample(10,random_state=101)

```

```python
# head the new df

merge_adult.head()
```

```python
# Merge data using common keys.
# create new df by subset of columns from orig df then sample 10 rows and seed the rand # gen

merge2_adult=adult_df[['education','sex','occupation']].sample(10,random_state=101)
```

```python
# head the new df

merge2_adult.head()
```

```python
# Merge data using common keys.
# create new df merge from two new dfs on occupation and an inner join droping the dups.

final_adult=pd.merge(merge_adult, merge2_adult, on ='occupation',how ='inner').drop_duplicates()
```

Create a series and practice basic arithmetic steps


```python
# Create a series and practice basic arithmetic steps

import pandas as pd  #import library

# Creating Series 1
series1_data = [7.3, -2.5, 3.4, 1.5]
series1_index = ['a', 'c', 'd', 'e']
series1 = pd.Series(series1_data, index=series1_index)  # create the pandas series

# Creating Series 2
series2_data = [-2.1, 3.6, -1.5, 4, 3.1]
series2_index = ['a', 'c', 'e', 'f', 'g']
series2 = pd.Series(series2_data, index=series2_index)  # create the pandas series

# Adding Series 1 and Series 2
addition_result = series1 + series2
print("Addition Result:")
print(addition_result)

# Subtracting Series 1 from Series 2
subtraction_result = series2 - series1
print("\nSubtraction Result:")
print(subtraction_result)
```
