```python
import http.client
import json
import pandas as pd
import time

teams_df = pd.DataFrame(teams_data)  # Assuming you already have the teams_df DataFrame

# Define positions to filter
positions_to_filter = ["QB", "WR", "RB", "TE"]

# Initialize a dictionary to store DataFrames for each year
yearly_dfs = {}

for index, row in teams_df.iterrows():
    team_id = row['home_id']
    print(f"Fetching statistics for team with ID: {team_id}")
    for year in range(2017, 2024):
        # Initialize lists for the current year
        player_ids = []
        player_names = []
        player_jersey_numbers = []
        player_positions = []

        try:
            conn = http.client.HTTPSConnection("api.sportradar.us")
            url = f"/nfl/official/trial/v7/en/seasons/{year}/REG/teams/{team_id}/statistics.json?api_key=k8uqtrmhauf3rmkm6f8t9f54"
            print("Requesting URL:", url)
            conn.request("GET", url)
            res = conn.getresponse()
            data = res.read()
            
            # Decode JSON response
            decoded_data = json.loads(data.decode("utf-8"))

            # Filter player statistics for specified positions
            for player in decoded_data["players"]:
                if player["position"] in positions_to_filter:
                    # Collect player information for the current year
                    player_ids.append(player["id"])
                    player_names.append(player["name"])
                    player_jersey_numbers.append(player["jersey"])
                    player_positions.append(player["position"])

            # Create a DataFrame for the current year and team
            player_df = pd.DataFrame({
                "player_id": player_ids,
                "name": player_names,
                "jersey_number": player_jersey_numbers,
                "position": player_positions
            })

            # Check if the year is already in the dictionary, if not, initialize a list for that year
            if year not in yearly_dfs:
                yearly_dfs[year] = []

            # Append the DataFrame to the list for the year
            yearly_dfs[year].append(player_df)

            # Introduce a delay of 3 seconds between requests to avoid rate limiting
            time.sleep(3)
        except Exception as e:
            print(f"Error fetching data for team ID {team_id} in year {year}: {e}")

# Now, yearly_dfs contains a list of DataFrames for each year, 
# where each DataFrame corresponds to a team's player data for that year.
# To access the DataFrames for a specific year, you can use yearly_dfs[year],
# and you can concatenate them if needed using pd.concat(yearly_dfs[year]).

```

```python
import http.client
import json
import pandas as pd

conn = http.client.HTTPSConnection("api.sportradar.us")

conn.request("GET", "/nfl/official/trial/v7/en/games/2021/reg/schedule.json?api_key=k8uqtrmhauf3rmkm6f8t9f54")

res = conn.getresponse()
data = res.read()

# Decode JSON response
decoded_data = json.loads(data.decode("utf-8"))

# Initialize an empty set to store unique home IDs
unique_home_ids = set()

# Initialize an empty list to store teams data
teams_data = []

# Extract distinct home IDs, names, and aliases along with venue details
for week in decoded_data["weeks"]:
    for game in week["games"]:
        home_id = game["home"]["id"]
        if home_id not in unique_home_ids:
            home_name = game["home"]["name"]
            home_alias = game["home"]["alias"]
            venue_id = game["venue"]["id"]
            venue_name = game["venue"]["name"]
            surface = game["venue"]["surface"]
            roof_type = game["venue"]["roof_type"]
            teams_data.append({
                "home_id": home_id,
                "name": home_name,
                "alias": home_alias,
                "venue_id": venue_id,
                "venue_name": venue_name,
                "surface": surface,
                "roof_type": roof_type
            })
            unique_home_ids.add(home_id)

# Create a DataFrame from the collected data
teams_df = pd.DataFrame(teams_data)

# Print the DataFrame
print(teams_df)
```
