```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup

all_stats = []  # To store all player statistics for all years

for year in range(2017, 2024):
    url = "https://www.pro-football-reference.com/years/{}/fantasy.htm".format(year)
    html = urlopen(url)
    soup = BeautifulSoup(html, 'html.parser')

    # Extracting headers from the second row of the table
    headers = [th.getText() for th in soup.findAll('tr')[1].findAll('th')]
    headers = headers[1:]  # Exclude the first (0 index) column header

    # Extracting player statistics from the table
    rows = soup.findAll('tr', class_=lambda table_rows: table_rows != "thead")
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
                    for i in range(len(rows))]
    player_stats = player_stats[2:]  # Exclude the first two rows as they might not contain player data

    # Create a DataFrame for the current year
    stats = pd.DataFrame(player_stats, columns=headers)

    # Print the first 5 player statistics for the year
    print(f"Player statistics for the year {year}:\n{stats.head()}")

    # Append the current year's statistics to the overall list
    all_stats.append(stats)

    # Rest of your code for data analysis goes here
    # ...

    # Close the HTML connection (optional but good practice)
    html.close()

# Concatenate all DataFrames into one DataFrame for all years
all_years_stats = pd.concat(all_stats, ignore_index=True)
```
