```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
year = 2019
url = "https://www.pro-football-reference.com/years/{}/games.htm".format(year) #string formatting to manipulate the url by a given year
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
headers = [th.getText() for th in soup.findAll('tr')[0].findAll('th')] #Find the second table row tag, find every table header column within it and extract the html text via the get_text method.
headers = headers[1:] #Do not need the first (0 index) column header
print(headers[:5])
```

```python
rows = soup.findAll('tr', class_ = lambda table_rows: table_rows != "thead") #Here we grab all rows that are not classed as table header rows - football reference throws in a table header row everyy 30 rows 
player_stats = [[td.getText() for td in rows[i].findAll('td')] # get the table data cell text from each table data cell
                for i in range(len(rows))] #for each row
player_stats = player_stats[2:]
```

```python
stats = pd.DataFrame(player_stats, columns = headers)
stats.head()
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

year = 2019
url = "https://www.pro-football-reference.com/years/{}/games.htm".format(year)
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')  # Specify the parser explicitly

# Print the HTML content to inspect the structure
print(soup.prettify())

# Find the table with the class 'suppress_all sortable stats_table'
table = soup.find('table', {'class': 'suppress_all sortable stats_table'})

# Check if the table exists before trying to extract headers
if table:
    # Extract headers from the first row of the table
    headers = [th.get_text() for th in table.find('thead').find_all('th')][1:]
    print(headers[:5])
else:
    print("Table not found.")

```
