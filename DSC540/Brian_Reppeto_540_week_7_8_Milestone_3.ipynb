```python
from selenium import webdriver
from bs4 import BeautifulSoup

# Set the path to your ChromeDriver executable
#chrome_driver_path = '/Users/brianreppeto/chromedriver'
chrome_driver_path = 'chromedriver'


# Set up the Selenium WebDriver
driver = webdriver.Chrome(executable_path=chrome_driver_path)

url = 'https://www.rotowire.com/betting/nfl/archive.php'
driver.get(url)

# Wait for the page to load (you may need to adjust the sleep time based on your network speed)
import time
time.sleep(5)

# Get the page source after it has been modified by JavaScript
html_content = driver.page_source

# Close the WebDriver
driver.quit()

# Parse the HTML with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Now you can work with the parsed HTML as before
table = soup.find('table')

if table:
    print(table.prettify())
else:
    print("Table not found.")
```

```python
if table:
    # Iterate over rows
    for row in table.find_all('tr'):
        # Iterate over cells in each row
        for cell in row.find_all(['th', 'td']):
            print(cell.text.strip(), end='\t')
        print()  # Move to the next line after each row
else:
    print("Table not found.")
```
